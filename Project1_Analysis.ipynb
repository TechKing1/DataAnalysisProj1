## We will now read the CSV

```python
import pandas as pd
df = pd.read_csv("sleep_gpa3.csv")
df.head()
```

```python
df.info()
```

## Bar chart of hours slept

```python
hours_freq = df['Sleep'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
plt.bar(hours_freq.index, hours_freq.values)
plt.xlabel('Hours slept')
plt.ylabel('Frequency')
plt.title('Frequency of hours slept')
plt.xticks(hours_freq.index)
plt.grid(axis='y')
plt.show()
```

## Bar chart of academic years

```python
year_freq = df['Year'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
plt.bar(year_freq.index, year_freq.values)
plt.xlabel('Academic Year')
plt.ylabel('Frequency')
plt.title('Frequency of academic years')
plt.xticks(year_freq.index)
plt.grid(axis='y')
plt.show()
```

## Bar chart of genders

```python
gender_freq = df['Gender'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(gender_freq.index, gender_freq.values)
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Frequency of Genders')
plt.grid(axis='y')
plt.show()
```

## Box plot of average GPA

```python
gp = df['CGPA']
mean2 = np.mean(gp)
median2 = np.median(gp)
mode2 = stats.mode(gp)

print("Mean:", mean2)
print("Median:", median2)
print("Mode:", mode2)

plt.figure(figsize=(8, 6))
plt.boxplot(gp, vert=False)
plt.xlabel('CGPA')
plt.title('Boxplot of the Average CGPA')
plt.show()
```

## Average GPAs per hours of sleep

```python
df.groupby('Sleep')['CGPA'].mean()
```

```python
data = {'5-6 hours':3.41, 'Less than 5 hours':3.40, 'More than 8 hours': 3.30,
        '6-7 hours':3.24, '7-8 hours':2.60}
hours = list(data.keys())
grades = list(data.values())
fig = plt.figure(figsize = (13, 9))
# creating the bar plot
plt.bar(hours, grades, color ='maroon', width = 0.4)
plt.xlabel("Hours slept")
plt.ylabel("Average CGPA")
plt.title("CGPAs per hours slept")
plt.show()
```

## Pie chart of faculties

```python
faculties = df['Faculty']
faculties_freq = faculties.value_counts()
plt.figure(figsize=(9.75, 15))
plt.pie(faculties_freq, labels=faculties_freq.index, autopct='%1.1f%%', startangle=90)
plt.title('Responses from different faculties')
plt.show()
```

## Pie chart of hours of sleep

```python
sleep = df['Sleep']
sleep_freq = sleep.value_counts()
plt.figure(figsize=(9.75, 15))
plt.pie(sleep_freq, labels=sleep_freq.index, autopct='%1.1f%%', startangle=90)
plt.title('Different hours of sleep')
plt.show()
```

## Histogram of GPAs

```python
from matplotlib import pyplot as plt
import numpy as np
# Creating dataset
a = np.array(df['CGPA'])
# Creating histogram
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(a, bins = [1.5, 2.5, 3.0, 3.5, 4.0])
# Show plot
plt.show()
```
